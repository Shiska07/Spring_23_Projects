```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## K Means

#### Helper Functions

```python
# reads a tab/whitespace separated file and returns a numpy array
def get_data_arr(file_path):
    
    # read as dataframe
    df = pd.read_csv(file_path, delim_whitespace = True, lineterminator='\n', header = None)
    
    # drop last column
    df.drop(columns = df.columns[-1], axis = 1, inplace = True)
    
    # convert dataframe to numpy arrar
    data_arr = df.to_numpy()
    
    # get number of features and samples
    n_samp, n_feat = data_arr.shape
    
    return data_arr, n_samp, n_feat


# computes centroids given cluster assignments
def calculate_cluster_centroids(data_arr, n_samp, n_feat, cluster_assignments):
    
    # get number of clusters
    n_c = max(cluster_assignments) + 1

    # dictionary to store new centroids; key: cluster_number, value: centroid
    centroids_dict = {}

    # compute centroids using cluster assignments
    for cluster_number in range(n_c):

        # use array slicing to get datapoints belonging to cluster
        x = data_arr[cluster_assignments == cluster_number]

        # if there are no datapoints assignd to the cluster
        if x.shape[0] != 0:
            centroids_dict[cluster_number] = (np.sum(x, axis=0) / x.shape[0]).reshape(1, n_feat)
        else:
            centroids_dict[cluster_number] = np.sum(x, axis=0).reshape(1, n_feat)

    return centroids_dict


# initalizes cluster centroids using random parition method
def initialize_centroids(data_arr, n_samp, n_feat, k):
    # dictionary to store centroids; key: cluster_number, value: centroid
    centroids_dict = {}

    # randomly assign datapoints to a cluster
    random_cluster_assignments = np.random.randint(low=0, high=k, size=(n_samp,), dtype=int)

    # return cluster centroids for random assignments
    return calculate_cluster_centroids(data_arr, n_samp, n_feat, random_cluster_assignments)


# gets new cluster assignments given centroids using euclidean distance
def update_cluster_assignments(data_arr, n_samp, n_feat, centroids_dict):
    # get number of clusters
    n_c = len(centroids_dict.keys())

    # array to store distance from each cluster
    dist_arr = np.zeros((n_samp, n_c), dtype=float)

    # caluclate euclidean distance from each centroid
    for cluster_number in range(n_c):
        # each column contains distance value from a cluster centroid
        # column index represents cluster_number
        dist_arr[:, cluster_number] = np.sqrt(np.sum((data_arr - centroids_dict[cluster_number]) ** 2, axis =1))

    # for each sample, index of the column of the lowest distance is closest cluster
    # and hence the cluster assignment
    cluster_assignments = np.argmin(dist_arr, axis = 1)

    return cluster_assignments


# calculates SSE error given final cluster assignments
def get_SSE(data_arr, n_samp, n_feat, cluster_assignments, centroids_dict):
    
    # compute centroids
    n_c = max(cluster_assignments) + 1

    SSE = 0

    for cluster_number in range(n_c):
        # use array slicing to get datapoints belonging to cluster
        x = data_arr[cluster_assignments == cluster_number]

        # compute euclidean distance
        dist_arr = np.sqrt(np.sum((x - centroids_dict[cluster_number]) ** 2, axis=1))

        # add sum of distances to sse
        SSE += np.sum(dist_arr)

    return SSE

# for plotting error
def plot_SSE(x, sse_values, title, xlabel, ylabel):
    
    plt.figure(figsize = (8, 4))
    plt.plot(x, sse_values, color = 'red')

    # add title and labels
    plt.title(title, fontdict = {'fontsize': 14})
    plt.xlabel(xlabel, fontdict = {'fontsize': 12})
    plt.ylabel(ylabel, fontdict = {'fontsize': 12})
    plt.xticks(k_values)
    plt.grid()
    
```

### Clustering algorithm

```python
# given an unlabeled data, returns k number of clusters
def get_kmeans_clusters(data_arr, n_samp, n_feat, k):
    # initalize cluster centroids
    cluster_centroids_dict = initialize_centroids(data_arr, n_samp, n_feat, k)

    # get new cluster assignments
    prev_cluster_assignments = update_cluster_assignments(data_arr, n_samp, n_feat, cluster_centroids_dict)

    # i to count iterations
    i = 0
    SSE_final = 0  # final SSE
    SSE_20 = 0  # SSE after 20th iteration

    while True:

        # compute centroids
        cluster_centroids_dict = calculate_cluster_centroids(data_arr, n_samp, n_feat, prev_cluster_assignments)

        if i == 19:
            # get error for 20th iteration
            # centroids_dict is updated centroids after cluster assignment
            SSE_20 = get_SSE(data_arr, n_samp, n_feat, prev_cluster_assignments, cluster_centroids_dict)

        # get new cluster assignments
        cluster_assignments = update_cluster_assignments(data_arr, n_samp, n_feat, cluster_centroids_dict)

        # if new assignments is same as the old assignments, stop
        if (prev_cluster_assignments == cluster_assignments).all():
            break

        # save new assignments
        prev_cluster_assignments = cluster_assignments

        # increment i
        i += 1

    # get error; centroids_dict is updated centroids after cluster assignment
    SSE_final = get_SSE(data_arr, n_samp, n_feat, cluster_assignments, cluster_centroids_dict)

    # return final cluster assignments, sse after 20th iteration and final SSE
    return cluster_assignments, SSE_20, SSE_final
```

### Satellite Data

```python
sat_data_path = 'UCI_datasets/satellite_training.txt'
```

```python
data_arr, n_samp, n_feat = get_data_arr(sat_data_path)
print(f"File '{sat_data_path}' has {n_samp} samples with {n_feat} features.")
```

#### Get SSE values for k = 2 to 10

```python
# initialize cluster numbers
k_values = list(np.arange(2, 11))

# list to store SSE values
SSE_values_20 = []          # after 20th iteration
SSE_values_final = []       # after final iteration

for k in k_values:
    __, sse_20, sse_final = get_kmeans_clusters(data_arr, n_samp, n_feat, k)
    SSE_values_20.append(sse_20)
    SSE_values_final.append(sse_final)
```

#### Print Results

```python
print(f"Results for file '{sat_data_path}':\n")
for i in range(len(SSE_values_20)):
    print(f'SSE for {k_values[i]} clusters after 20 iterations is {SSE_values_20[i]:0.4f}.')  
```

```python
print(f"Results for file '{sat_data_path}':\n")
for i in range(len(SSE_values_final)):
    print(f'SSE for {k_values[i]} clusters after convergence is {SSE_values_final[i]:0.4f}.')
```

### Plot Results

```python
title = 'SSE vs k after 20 iterations for Satellite Dataset'
xlabel = 'Number of clusters(k)'
ylabel = 'SSE'

plot_SSE(k_values, SSE_values_20, title, xlabel, ylabel)
```

```python
title = 'SSE vs k after convergence for Satellite Dataset'
xlabel = 'Number of clusters(k)'
ylabel = 'SSE'

plot_SSE(k_values, SSE_values_final, title, xlabel, ylabel)
```
